```python
#histplot

import matplotlib.pyplot as plt
import pandas as pd

df['Publishing Year']=df['Publishing Year'].abs()
df_1=df[df['Publishing Year'] >= 1900]
plt.figure(figsize =(8,5))
plt.hist(df_1['Publishing Year'], bins =15, color='orange', edgecolor='black')
plt.title('Histogram of Publishing Year')
plt.xlabel('Publishing Year')
plt.ylabel('No. of books')
plt.yticks([25,50,75,100,125,150,175,200,225,250,275,300])
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_1.columns = df_1.columns.str.strip().str.lower()

#  Create box plot
plt.figure(figsize=(10,6))
sns.boxplot(x='genre', y='book_ratings_count', data=df_1, hue='genre', palette='Set3', legend=False)

plt.title('Book Ratings Count by Genre', fontsize=14)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Book Ratings Count', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get language counts and convert to percentage
language_counts = df_1['language_code'].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 5))
language_counts.sort_values().plot(kind='barh', color='skyblue', edgecolor='black')

# Add percentage labels to bars
for index, value in enumerate(language_counts.sort_values()):
    plt.text(value + 0.5, index, f'{value:.1f}%', va='center', fontsize=9)

plt.title('Language Distribution of Books', fontsize=14)
plt.xlabel('Percentage of Books (%)', fontsize=12)
plt.ylabel('Language Code', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get value counts as percentages
language_count = df_1['language_code'].value_counts(normalize=True) * 100

# Convert values to floats for plotting, and round for display
language_count = language_count.sort_values()
percent_values = language_count.values
labels = language_count.index.astype(str)  # Ensure labels are strings

# Plot line chart
plt.figure(figsize=(8,5))
plt.plot(labels, percent_values, marker='o', linestyle='-', color='teal')

# Add value labels above each point
for x, y in zip(labels, percent_values):
    plt.text(x, y + 0.5, f'{y:.1f}%', ha='center', fontsize=9)

plt.title('Language Distribution of Books (Line Chart)', fontsize=14)
plt.xlabel('Language Code', fontsize=12)
plt.ylabel('Percentage of Books (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by author_rating and get max book_ratings_count
max_rating_count = df_1.groupby('author_rating')['book_ratings_count'].max().sort_index()

# Plot as line chart
plt.figure(figsize=(8,5))
plt.plot(max_rating_count.index, max_rating_count.values, marker='o', linestyle='-', color='darkgreen')

plt.title('Max Book Ratings Count by Author Rating', fontsize=14)
plt.xlabel('Author Rating', fontsize=12)
plt.ylabel('Max Book Ratings Count', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#PIE CHART : Share of Books by author_rating
import matplotlib.pyplot as plt

# Count books per author rating
author_rating_counts = df_1['author_rating'].value_counts().sort_index()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(
    author_rating_counts,
    labels=author_rating_counts.index,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140,
    colors=plt.cm.Pastel2.colors,
    textprops={'fontsize': 10}
)

plt.title('Book Distribution by Author Rating', fontsize=14)
plt.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.figure(figsize=(8,5))
plt.scatter(
    df_1['book_ratings_count'],
    df_1['book_average_rating'],
    color='red',
    alpha=0.4,
    edgecolors='black'
)

plt.title('Book Ratings Count vs. Average Rating', fontsize=14)
plt.xlabel('Book Ratings Count', fontsize=10)
plt.ylabel('Book Average Rating', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.1)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by author, sum gross sales, and get top 5
top5_authors = df_1.groupby('author')['gross sales'].sum().sort_values(ascending=False).head(5)

# Plot bar chart
top5_authors.plot(kind='bar', figsize=(8,4), color='skyblue', edgecolor='black')

plt.title('Top 5 Authors by Gross Sales')
plt.xlabel('Author')
plt.ylabel('Gross Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# box plot for author rating and units sold using seaborn
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Clean column names (if not already done)
df_1.columns = df_1.columns.str.strip().str.lower()

# Step 2: Create box plot
plt.figure(figsize=(8,6))
sns.boxplot(x='author_rating', y='units sold', data=df_1,hue='author_rating', palette='Set2',
    legend=False)

plt.title('Units Sold by Author Rating', fontsize=14)
plt.xlabel('Author Rating', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
yearly_sales = df_1.groupby('publishing year')['units sold'].sum().sort_index()

# Step 3: Plot line chart
plt.figure(figsize=(10,5))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', linestyle='-', color='darkblue')

plt.title('Units Sold by Publishing Year', fontsize=14)
plt.xlabel('Publishing Year', fontsize=12)
plt.ylabel('Total Units Sold', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
